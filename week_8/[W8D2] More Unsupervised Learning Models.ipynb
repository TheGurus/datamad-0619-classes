## 0. python imports

```python
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats

%matplotlib inline
set_matplotlib_formats('svg')
```

## 1. data loading & generation

```python
from sklearn.datasets import make_moons

moons = make_moons(n_samples=100, 
                   shuffle=True, 
                   noise=0.05)

plt.scatter(moons[0][:,0], moons[0][:,1], c=moons[1]);
```

## 2a. DBSCAN (density based) clustering

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.2)
dbscan.fit(moons[0]);
```

```python
plt.scatter(moons[0][:,0], moons[0][:,1], c=dbscan.labels_);
```

## 2b. HDBSCAN (hierachical density based)

```python
from hdbscan import HDBSCAN
hdbscan = HDBSCAN()
hdbscan.fit(moons[0]);
```

```python
plt.scatter(moons[0][:,0], moons[0][:,1], c=hdbscan.labels_);
```

```python
plt.scatter(moons[0][:,0], moons[0][:,1], c=hdbscan.probabilities_);
```

## 3. GMM (gaussian mixture models)

```python
from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=2, covariance_type='full')
gm.fit(moons[0]);
```

```python
gm.means_
```

```python
gm.covariances_
```

```python
gm.predict_proba(moons[0])
```

```python
plt.scatter(moons[0][:,0], moons[0][:,1], c=gm.predict_proba(moons[0])[:,1]);
```

```python
plt.scatter(moons[0][:,0], moons[0][:,1], c=gm.predict(moons[0]));
```

```python
from sklearn.datasets import make_classification

X = make_classification(n_features=2, 
                        n_redundant=0, 
                        n_informative=2,
                        n_clusters_per_class=1)

gm = GaussianMixture(n_components=2, covariance_type='full')
gm.fit(X[0]);
```

```python
plt.scatter(X[0][:,0], X[0][:,1], c=gm.predict_proba(X[0])[:,1]);
```

```python
plt.scatter(X[0][:,0], X[0][:,1], c=gm.predict(X[0]));
```
