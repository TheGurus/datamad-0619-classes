```python
import matplotlib.pyplot as plt
import pandas as pd

from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')

pd.options.display.max_columns = None
```

```python
census = pd.read_csv('./acs2015_county_data.csv')
census.head()
```

```python
census.describe()
```

```python
census.isnull().sum(axis = 0)
```

```python
census_missing = census.dropna()
```

```python
census_columns = [col for col in census.columns.values if col not in ['CensusId', 'State', 'County']]
```

# K-Means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
```

```python
census_clusters = kmeans.fit(census_missing[census_columns])

# really difficult to plot (34-dimensional space)
census_clusters.cluster_centers_
```

```python
# predictions of to which cluster the observation belongs
predictions = census_clusters.fit_predict(census_missing[census_columns])
predictions
```

```python
# distances to clusters for each observation
distances = census_clusters.fit_transform(census_missing[census_columns])
distances
```

```python
import numpy as np

# for those in cluster 1, the minimum distance is cluster 1 distance
distances[np.where(predictions == 1)[0]]
```

```python
census_missing['Cluster'] = predictions
```

```python
census_missing[census_missing['Cluster'] == 2]
```

```python
census_missing[census_missing.Cluster == 0].State.value_counts()
```

```python
census_missing.groupby(['Cluster'])['Income'].mean()
```
